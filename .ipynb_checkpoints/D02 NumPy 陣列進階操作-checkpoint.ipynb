# NumPy 陣列重塑 - flatten() 與 ravel()


```python
arr.flatten()
```

```python
arr.flatten()[5] = 100
arr
```

註：介於**call-by-value**和**call-by-reference**之間

```python
arr.ravel()[5] = 100
arr
```

# NumPy 陣列的合併 - stack(), hstack(), vstack()


```python
arr_1 = np.array([[1, 3, 5], [2, 4, 6]])
arr_1
```

```python
arr_2 = np.array([[11, 13, 15], [12, 14, 16]])
arr_2
```

```python
# stack
# 一個array自成一個維度
# 所以每個array的長相都必須一樣
np.stack([arr_1, arr_2])
```

```python
np.hstack([arr_1, arr_2])
```

```python
np.vstack([arr_1, arr_2])
```

# NumPy 陣列的分割 – split()、hsplit()、vsplit()


```python
arr_3 = np.arange(1, 13, 1).reshape((6, 2))
arr_3
```

```python
np.split(arr_3, 2)
```

```python
np.split(arr_3, [2, 4 ,6])
# 相當於
# arr_3[:2]
# arr_3[2:4]
# arr_3[4:6]
```

```python
np.hsplit(arr_3, 2)
# 水平軸切成2個dimension
```

```python
np.vsplit(arr_3, 2)
# 垂直軸切成2個dimension
```

# NumPy 陣列迭代


```python
for row in arr_3:
    print(row)
```

```python
for element in arr_3.flat:
    print(element)
```

# NumPy 陣列搜尋與排序 - amax()、amin()、max()、min()


```python
arr_4 = np.array([[10, 1], [1, 11], [1, 12]])
arr_4
```

```python
np.amax(arr_4)
```

```python
arr_4.max()
```

```python
arr_4.max(axis=0)
```

```python
arr_4.max(axis=1)
```

```python
arr_3.max(axis=0, keepdims=True)
```

# NumPy 陣列搜尋與排序 – argmax() 與 argmin()


```python
np.argmax(arr_4)
```

```python
arr_4.argmax()
```

```python
arr_4.argmax(axis=0)
```

```python
arr_4.argmax(axis=1)
```

# NumPy 陣列搜尋與排序 – where()


```python
arr_5 = np.linspace(11, 20, 10)
arr_5
```

```python
np.where(arr_5 >= 15)
```

```python
arr_6 = np.linspace(11, 20, 10).reshape((5, 2))
arr_6
```

```python
np.where(arr_6 >= 15)
```


表`arr_6[2, 0]`, `arr_6[2, 1]`...等滿足條件
